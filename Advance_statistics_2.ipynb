Question 6: What is the Central Limit Theorem (CLT), and why is it critical in statistics?
Answer:  
The Central Limit Theorem is a fundamental principle in statistics that states:
When independent random variables are added, their normalized sum tends toward a normal distribution, regardless of the original distribution, as the sample size becomes large.

   Example:
Suppose individual weights of apples follow a skewed distribution, but if you take the average weight of 50 apples repeatedly, the distribution of those averages will be approximately normal due to the CLT.




Question 7: What is the significance of confidence intervals in statistical analysis?
Answer:  
 A confidence interval (CI) is a range of values, derived from sample data, that is likely to contain the true population parameter (such as mean or proportion) with a certain level of confidence.
Confidence Interval=Point EstimateÂ±Margin of Error\text{Confidence Interval} = \text{Point Estimate} \pm \text{Margin of Error}Confidence Interval=Point EstimateÂ±Margin of Error
Significance of Confidence Intervals:
1.	Estimate population parameters:
o	CIs provide a range in which the true value (mean, proportion, etc.) is likely to lieâ€”not just a single number.
2.	Account for sampling variability:
o	Since samples can vary, CIs express the uncertainty in estimates.
3.	Communicate reliability:
o	A narrower interval = more precise estimate (usually from larger samples or less variability).
o	A wider interval = less precise.
4.	Support decision-making:
o	Helps determine statistical significanceâ€”e.g., if a CI for a difference excludes 0, it may imply a significant effect.
5.	Used in hypothesis testing:
o	If a CI does not contain the null hypothesis value (e.g., mean = 0), the result is statistically significant at that confidence level.

```python
'''Question 9: Write a Python program to generate 1000 random numbers from a normal distribution with mean = 50 and standard deviation = 5. Compute its mean and standard deviation using NumPy, and draw a histogram to visualize the distribution. (Include your Python code and output in the code box below.)  '''

import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate 1000 random numbers from normal distribution
data = np.random.normal(loc=50, scale=5, size=1000)

# Step 2: Compute mean and standard deviation
mean = np.mean(data)
std_dev = np.std(data)

# Step 3: Print results
print(f"Mean of the generated data: {mean:.2f}")
print(f"Standard Deviation of the generated data: {std_dev:.2f}")

# Step 4: Plot histogram
plt.figure(figsize=(8, 5))
plt.hist(data, bins=30, color='skyblue', edgecolor='black', density=True)
plt.title('Histogram of Normally Distributed Data')
plt.xlabel('Value')
plt.ylabel('Frequency Density')
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean = {mean:.2f}')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
